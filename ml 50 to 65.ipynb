```python
# how to use and & or operator
import pandas as pd  #till now whatever we wrote now we are doing in 3 lines  of code
df=pd.read_csv('employees.csv',parse_dates=['start date','last login time'])
df['senior manegement']=df['senior manegement'].astype('bool')
df['gender']=df['gender'].astype('category')
df.head()
```

```python
mask1=df['gender']=='male'
mask2=df['Team']=='marketing'
df[mask1 & mask2]
```

```python
mask3=df['senior manegement']
mask4=df['gender']=='male'
df[mask3 | mask4]
```

```python
mask5=df['Team']=='marketing'
mask6=df['senior manegement']
mask7=df['gender']=='male'
df[mask5 | (mask6 & mask7)]
```
