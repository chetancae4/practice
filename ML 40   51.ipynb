```python
#rank methode
import pandas as pd
nba=pd.read_csv('nba.csv')
nba.head()
```

```python
nba['salary']=nba['salary'].fillna(0).astype(int)    # as type means o is type of int
nba
```

```python
nba['salary'].rank().astype(int)  #rank is in ascending orer means who has lowest salary will get 1 st rank
# ex:---index (7) i'e rajs has zero salary so his rank is 1st were as hars has highest salary so her rank is last i'e 10th
# 0  9   means crs has 9th rank
```

```python
nba['salary rank']=nba['salary'].rank(ascending=False).astype(int)
nba
```

```python
nba.sort_values("salary rank")
```
