```python
#aply methodeimport pandas as pd 
import pandas as pd
bond=pd.read_csv('jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head()
```

```python
def add_millions(number):
    return str(number)+'millions'
bond['Box Office'].apply(add_millions)
```

```python
bond['Box Office']=bond['Box Office'].apply(add_millions)
```

```python
def add_millions(number):
    return str(number)+' millions'
bond['Budget'].apply(add_millions).head()
```

```python
def add_millions(number):
    return str(number)+' millions'
columns=['Box Office','Budget','Bond Actor Salary']
for col in columns:
    bond[col]=bond[col].apply(add_millions)
bond.head()
```

```python
#how to use apply methode row wise
def good_movie(row):
    actor=row[1]
    budget=row[4]
    if actor == 'Pierce Brosnan':
        return 'The Best'
    elif actor=='Roger Moore' and budget > 40:
        return 'engoyable'
    else:
        return 'i have no idea'
bond.apply(good_movie,axis='columns')
    
```
